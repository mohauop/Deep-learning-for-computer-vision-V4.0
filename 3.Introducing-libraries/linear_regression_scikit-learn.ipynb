```python
import numpy as np 
import matplotlib.pyplot as mtp  
import pandas as pd 
```

```python
url = 'https://raw.githubusercontent.com/AshishJangra27/Machine-Learning-with-Python-GFG/main/Linear%20Regression/data_for_lr.csv'
data = pd.read_csv(url)
data = data.dropna()
```

```python

# training dataset and labels
train_input = np.array(data.x[0:500]).reshape(500, 1)
train_output = np.array(data.y[0:500]).reshape(500, 1)

# valid dataset and labels
test_input = np.array(data.x[500:700]).reshape(199, 1)
test_output = np.array(data.y[500:700]).reshape(199, 1)
```

## check this website
https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
#Fitting the Simple Linear Regression model to the training dataset  
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression()  
regressor.fit(train_input, train_output)  
```

```python
#Prediction of Test and Training set result  
y_pred= regressor.predict(test_input)  
x_pred= regressor.predict(train_input)  
```

```python
mtp.scatter(train_input, train_output, color="green")   
mtp.plot(train_input, x_pred, color="red")    
mtp.title("Salary vs Experience (Training Dataset)")  
mtp.xlabel("Years of Experience")  
mtp.ylabel("Salary(In Rupees)")  
mtp.show()   
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(test_output, y_pred))
```

## Evaluation metrics

```python
def mae_metric(actual, predicted):
    sum_error = 0.0
    for i in range(len(actual)):
        sum_error += abs(predicted[i] - actual[i])
    return sum_error / float(len(actual))
def mse_metric(actual, predicted):
    return np.mean((actual - predicted) ** 2)
def rmae_metric(mae):
    return np.sqrt(mae)
```

```python
mse = mse_metric(test_output, y_pred)
mae = mae_metric(test_output, y_pred)
rmae = rmae_metric(mae)
```

```python
print(f"MSE : {mse }")
print(f"MAE : {mae}")
print(f"RMAE : {rmae }")
```
